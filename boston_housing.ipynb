```python
X = boston.data
Y = boston.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
# Model 

from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
#print the coefficients and the intercept

model.coef_
```

```python
model.intercept_
```

```python
##on which parameters the model has been trained
model.get_params()
```

```python
model_predict = model.predict(X_test)
```

```python
model_predict
```

```python
#scatter plot for the prediction

plt.scatter(model_predict, y_test)
```

```python
#residuals
residuals = y_test- model_predict
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
#scatter plot wrt prediction and residuals
#uniform distribution
plt.scatter(model_predict, residuals)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, model_predict)
```

```python
np.sqrt(mean_squared_error(y_test, model_predict))
```

```python
# R squared
from sklearn.metrics import r2_score
r2_score(y_test, model_predict)
```
